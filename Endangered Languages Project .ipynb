```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline


#task 1 getting the data set
df = pd.read_csv('data.csv')
df
```

```python
#task 2:Basic Visualizations 

X = df["Degree of endangerment"].unique()
Y = pd.value_counts(df["Degree of endangerment"])
  
# Plot the data using bar() method
plt.bar(X, Y,width = 0.8,align = 'center', color=['green','yellow','orange','brown','purple'])

  
# Show the plot

plt.title('Endengerd languages')
plt.xlabel('Degree of endangerment')
plt.ylabel('Number of Languages')
plt.xticks(rotation = 90)
plt.show()
```

```python
#Task 3: East African Languages
selection = ["Kenya", "Ethiopia","Uganda", "Sudan","Somalia", "Rwanda","Burundi"]
sv = df[df.Countries.isin(selection)]
Y = pd.value_counts(sv["Degree of endangerment"])
X = df["Degree of endangerment"].unique()

# Plot the data using bar() method
plt.bar(X,Y, width = 0.8,align = 'center',color = ['green','yellow','orange','brown','purple'])

# Show the plot

plt.xlabel("Degree of Enadangerment")
plt.ylabel("Number of languages")
plt.title("East Africa Endengerd languages")
plt.xticks(rotation = 90)


```
